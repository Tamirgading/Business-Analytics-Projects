```python
import pandas as pd

df = pd.read_csv("final_shop_6modata.csv")
df.sample(5)
```

```python
df['Ad Group'] = df['Ad Group'].astype('string')
df.dtypes
```

```python
df[['Ad Type', 'Device Type', 'Campaign Type']] = df['Ad Group'].str.extract(r'Shop\s-\s(1:1|Exact|Phrase)\s-\s(Desk|Mob)\s-\s(.+)')
df['Campaign Type'] = df['Campaign Type'].str.replace(r'\[|\]', '', regex=True).str.strip()
df['Campaign Type'] = df['Campaign Type'].str.title()
df[['Ad Group', 'Ad Type', 'Device Type', 'Campaign Type']].head()
```

```python
df.head()
```

```python
df['Device Type'][df['Device Type'].str.contains('Desk')] = 'Desktop'
df['Device Type'][df['Device Type'].str.contains('Mob')] = 'Mobile'
```

```python
df['CPM'] = df['Cost'] / df['Impressions'] * 1000
df['CPM'] = df['CPM'].round(2)
df[['Impressions', 'Cost', 'CPM']].head()
```

```python
df.to_csv("rivised_final_shop_6modata.csv")
```
